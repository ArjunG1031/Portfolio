## Homework
Make a nested conditional that does the following: 
1. contains more than 1 nested conditional
2. use the numbers 1-50
3. check if each number is divisible by each factor of 50

```python
%%js
// CODE_RUNNER: Nested Conditonals Hw

for (let i = 1; i <= 50; i++) {
  console.log("Number:", i);

  if (i % 1 === 0) {
    console.log("  Divisible by 1");

    if (i % 2 === 0) {
      console.log("  Divisible by 2");
    }

    if (i % 5 === 0) {
      console.log("  Divisible by 5");
    }

    if (i % 10 === 0) {
      console.log("  Divisible by 10");
    }

    if (i % 25 === 0) {
      console.log("  Divisible by 25");
    }

    if (i % 50 === 0) {
      console.log("  Divisible by 50");
    }
  }
}
```
